```
from google.colab.patches import cv2_imshow
from google.colab import drive 

drive.mount('/content/drive')

#importar librerias
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
#graficar imagenes 
%matplotlib inline

```

```
imagenes =[]

image_dir='/content/drive/My Drive/Colab Notebooks/Imagen/cubo'

def get_image(imagen_path):
  image = cv2.imread(imagen_path)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  return image

for file in os.listdir(image_dir):
  if not file.startswith('.'):
    imagenes.append(get_image(os.path.join(image_dir, file)))

plt.figure(figsize=(20,10))
for i in range(len(imagenes)):
  plt.subplot(1, len(imagenes), i+1) 
  plt.imshow(imagenes[i])

def mostrar4(a,b,c,d, titulo1 ="original", titulo2="original",titulo3="original",titulo4="original"):
   plt.figure(figsize=(14,14))
   plt.subplot(221), plt.imshow(a), plt.title(titulo1)
   plt.xticks([])
   plt.yticks([])
   plt.subplot(222), plt.imshow(b,cmap='gray'), plt.title(titulo2)
   plt.xticks([])
   plt.yticks([])
   plt.subplot(223), plt.imshow(c,cmap='gray'), plt.title(titulo3)
   plt.xticks([])
   plt.yticks([])
   plt.subplot(224), plt.imshow(d, cmap='gray'), plt.title(titulo4)
   plt.xticks([])
   plt.yticks([])
   plt.show();
  

  
```

```
!pwd
img_BGR = cv2.imread('/content/drive/My Drive/Colab Notebooks/Imagen/cubo/rub07.jpg', cv2.IMREAD_COLOR)

b = img_BGR[:,:,0]
g = img_BGR[:,:,1]
r = img_BGR[:,:,2]

mostrar4(img_BGR,b,g,r,"original","b","g","r")
```

```
!pwd
img_RGB = cv2.cvtColor(img_BGR, cv2.COLOR_BGR2RGB)
r,g,b=cv2.split(img_RGB); 

mostrar4(img_RGB,r,g,b,"original","r","g","b")
```

```
count =2

plt.figure(figsize=(12,4))
plt.subplot(1,4,1),plt.imshow(img_RGB/255), plt.title("original"),plt.axis("off")

for i, lab in enumerate(["R","G","B"]):
  crgb = np.zeros(img_RGB.shape)
  crgb[:,:,i] = img_RGB[:,:,i]
  plt.subplot(1,4,count),plt.imshow(crgb/255),plt.title(lab)
  plt.axis("off")
  count +=1

plt.show()
```
